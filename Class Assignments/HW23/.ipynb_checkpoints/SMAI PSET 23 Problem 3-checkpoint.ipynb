```python
# Loading the MNIST data
# Using the test dataset provided for Assignment 1
# It has 100 samples of each digit

from __future__ import print_function
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.colors as mcolors

def read_data(filename):
    with open(filename, 'r') as f:
        lines = f.readlines()
    
    num_points = len(lines)
    dim_points = 28 * 28
    data = np.empty((num_points, dim_points))
    labels = np.empty(num_points)
    
    for ind, line in enumerate(lines):
        num = line.split(',')
        labels[ind] = int(num[0])
        data[ind] = [ int(x) for x in num[1:] ]
        
    return (data, labels)

data, labels = read_data("sample_test.csv")
```

```python
# Clearly, our dataset has 100 samples for each label
labels_count = np.zeros(10)
for label in labels:
    labels_count[int(label)] += 1
print("Number of Labels of each digit:", labels_count)
```

```python
# PCA to reduce the dataset to 2-dimensions
eigenvalues, eigenvectors = np.linalg.eig(np.cov(data.T))
length = np.shape(data)[0]

x = np.zeros(length)
y = np.zeros(length)
points = []

positions = np.flip(np.argsort(eigenvalues))

for iter in range(length):
    x[iter] = np.real(np.dot(data[iter,:], eigenvectors[:, positions[0]]))
    y[iter] = np.real(np.dot(data[iter,:], eigenvectors[:, positions[1]]))
    points.append((x[iter], y[iter]))
    
color_map = {0: 'red', 1: 'blue', 2: 'darkgreen', 3: 'yellow', 4: 'lime',
             5: 'navy', 6: 'brown', 7: 'orange', 8: 'darkviolet', 9: 'aqua'}

colors = [color_map[int(i)] for i in labels]
    
plt.figure()
plt.title('Dataplot after PCA')
plt.xlabel('x')
plt.ylabel('y')
plt.scatter(x, y, c = colors)
plt.show()
```

```python
from sklearn.cluster import KMeans
cluster_maker = KMeans(n_clusters=10, n_init=10, max_iter = 10000)
model = cluster_maker.fit(points)
print("Inertia:", model.inertia_)
```

```python
# Showing results of K-means clustering
colors = [color_map[int(i)] for i in model.labels_]
    
plt.figure()
plt.title('Dataplot after K-means clustering')
plt.xlabel('x')
plt.ylabel('y')
plt.scatter(x, y, c = colors)
plt.show()
```

```python
# We can use completeness_score as our measure of goodness
# completeness_metric gives a measure between 0 and 1, 1 being a perfect score
# when all members of a given class are assigned to the same cluster
# So we define our goodness_score as (completeness_metric * 100)

# Alternative Goodness_Score = homogeneity_metric * 100
# where homogeneity_metric captures how accurately each cluster contains only members of a single class
# But using completeness in this regard makes more intuitive sense

# For more on mathematical foundations of completeness & homogeneity scores, refer:
# https://scikit-learn.org/stable/modules/clustering.html#homogeneity-completeness

from sklearn.metrics import completeness_score

goodness_scores = {}

def goodness(ground_truth, results):
    score = completeness_score(labels, model.labels_) * 100
    return score

model_goodness = goodness(labels, model.labels_)
print("Goodness Measure of our Clustering:", model_goodness)
goodness_scores['k-means++ Initialization with (n_inits = 10)'] = model_goodness
```

```python
# Clustering of K-means depends on initialization
# So far we used the default 'k-means++' initialzation
# Now, we'll try 5 different random initializations

scores = []
centroids = []

fig = plt.figure(figsize=(40,20))

for index in range(6):
    cluster_maker_random = KMeans(n_clusters=10, init='random', n_init=1, max_iter=10000)
    model_random = cluster_maker_random.fit(points)
    results = model_random.labels_
    score = goodness(labels, results)
    scores.append(score)
    
    colors = [color_map[int(i)] for i in model_random.labels_]
    subplot_number = 230 + (index + 1)
    a = fig.add_subplot(subplot_number)
    a.scatter(x, y, c = colors)
    
    #Cluster Centre with given random initialization: model_random.cluster_centers_
    centroids.append(model_random.cluster_centers_)
    print("Goodness Metric with Random Initializations #:" + str(index + 1) + ": " + str(score))
    goodness_scores['Random Initialization #' + str(index + 1)] = score

plt.show()
```

```python
def compute_centroid(point_list):
    x_sum = y_sum = 0
    for point in point_list:
        x_sum += point[0]
        y_sum += point[1]
        
    return (x_sum / len(point_list), y_sum / len(point_list))

centroids = [] # Stores the centroid of each class

for label in range(10):
    label_points = [points[i] for i in range(length) if int(labels[i]) == label]
    centroids.append(compute_centroid(label_points))
    
print("The Centroids of Each Class:", centroids)
```

```python
# K-means initialized with ground truth
ground_truth_init = np.asarray(centroids)
cluster_maker_GT = KMeans(n_clusters=10, init=ground_truth_init, n_init=1, max_iter=10000)
model_GT = cluster_maker_GT.fit(points)
results = model_GT.labels_
score = goodness(labels, results)
print("Goodness Metric of K-means initialized with Ground Truth information:", str(score))
goodness_scores['Initialization with Ground Truth Information'] = score
print("Number of iterations to reach convergence:", str(model_GT.n_iter_))

# Clustering using K-means initialized with Ground Truth Labels
colors = [color_map[int(i)] for i in model_GT.labels_]
    
plt.figure()
plt.title('Dataplot after K-means clustering initialized with Ground Truth Information')
plt.xlabel('x')
plt.ylabel('y')
plt.scatter(x, y, c = colors)
plt.show()
```

```python
# Using Ground Truth for initialization
# K-means obviously does not converge to ground truth
# Ground truth information is never used for optimization
# Hence, clustering takes place as one would normally expect
# Convergence does not take place 'immediately', it requires 11 iterations
```

```python
# As we can see from the plots using random initializations
# The clustering does depend on the initialization
# There are some minor perceivable changes in the cluster structures
# However, the Goodness Score at the end of convergence is pretty much the same
# for all types of initializations

from IPython.display import display
import pandas as pd

index_arr = goodness_scores.keys()
d = {'Goodness Score:' : list(goodness_scores.values())}
df = pd.DataFrame(index = index_arr, data=d)
display(df)

# Here we see the goodness score values for all initializations
```
